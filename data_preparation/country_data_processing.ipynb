```python
import pandas as pd
```

#### Getting most recent data from world bank

- Population, GDP, Land Area data for different countries
- This datasets contains data from 1960
- Read from csv, converting into dataframe and select the recent data

Population data is collected from [https://data.worldbank.org/indicator/SP.POP.TOTL] site

```python
population = pd.read_csv("population.csv")
population_df = pd.DataFrame(population)
#print(population_df.columns)
population_df = population_df[["Country Name", "2019"]]
population_df = population_df.rename(columns={"2019":"population"})
population_df.head(5)
```

Gdp Data is collected from [https://data.worldbank.org/indicator/NY.GDP.MKTP.CD] site

```python
gdp = pd.read_csv("gdp.csv")
gdp_df = pd.DataFrame(gdp)
gdp_df = gdp_df[["Country Name", "2019"]]
gdp_df = gdp_df.rename(columns={"2019":"gdp"})

gdp_df.head(5)
```

Total land area in sq. meter Data is collected from [https://data.worldbank.org/indicator/AG.LND.TOTL.K2] site

```python
land_area = pd.read_csv("land_area.csv")
land_area_df = pd.DataFrame(land_area)
land_area_df = land_area_df[["Country Name", "2018"]]
land_area_df = land_area_df.rename(columns={"2018":"land_area"})
land_area_df.head(5)
```

### Merging data based on the country name (inner join)
##### and create a new dataset of 4 columns (country name, gdp, land_area and population)

```python
country_data = gdp_df.merge(land_area_df, on='Country Name').merge(population_df, on="Country Name")
```

```python
country_data.head()
```

```python
print(country_data.shape)
country_data["Country Name"].nunique()
#this datset has 264 unique country_name
```

### Covid-Time Financial Policy response data
#### The data is from wolrd bank [https://datacatalog.worldbank.org/dataset/covid-19-finance-sector-related-policy-responses]
#### We can join them by region or by date
#### we can track the policy change due to covid
#### There s also a option for NLP at Details of the measure column

```python
finance = pd.read_excel("covid-fci-data.xlsx")
finance_df = pd.DataFrame(finance)
finance_df.head()
```

#### Now lets join Finance Dataframe with coutry_Data based on country name
###### As all the data from World Name coutnry name should be consistent interms of spelling

```python
country_df_final = country_data.merge(finance_df, on="Country Name") 
```

```python
country_df_final.head()
```

#### we found data is not merged, and further inspection found that in finace data country name it has extra space,
#### we have to remove the space from the country name string

```python
#### custom funciton for removing the space
def remove_space(x):
      return x.replace(" ", "")
```

```python
finance_df["Country Name"] = finance_df["Country Name"].apply(remove_space)
```

```python
country_df_final = country_data.merge(finance_df, on="Country Name") 
country_df_final.head()
```

```python
country_df_final.to_csv("country_data.csv")
```
