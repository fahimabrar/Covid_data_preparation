##### the dataset is is 2020 air passenger traveling dataset
##### data source is collected from https://zenodo.org/record/4485741

```python
import pandas as pd
import glob
import datetime
#this will bring all the file that ends with csv.gz
air_data_list = glob.glob("*.csv.gz")
print(air_data_list)
airport_name = pd.read_csv("airport_data/airport.csv")
#airport dataset to match the destination Icao code code to a proper country name
```

```python
#### custom funciton for formatting the date as same as the covid dataset
#### we will need it later
def date_format(x):
    date = x.split()[0]
    return datetime.datetime.strptime(date, "%Y-%m-%d").strftime("%d/%m/%Y")
```

```python
#first giving explanation of the coding for a single month, 
# then we will run a for loop and do not explains much
data = pd.read_csv(air_data_list[0])
data.head()
```

```python
#selecting only the part we needed
airport_df = airport_name.iloc[:, 1:]
#assigning some nice column name, so that we can merge it with air passenger data and also with our covid data
airport_df.columns =['Country Name', 'Code', 'destination'] 
```

#### Icao code to destination country Name
- New problem occured
- this dataset have lattitude and longidute data for airport
- In our covid data we have lat long for the city name
- And city lat long does not match exactly with the airport lat long
- we tried using python library that reverse lat long to address
- the problem with the library is gives name in different language
- we searched a new dataset that have airport ICAO code and the coutnry of the airport
- we get a dataset from this website     https://openflights.org/data.html


```python
#join the data with the air passenger data to get the country name of the airport code Icao code(airport code) 
merged_passenger_df = data.merge(airport_df, on="destination", how="left")

#data.shape
data.head()
```

```python
#merged dataset looks like this
merged_passenger_df
```

```python
#to see if any airport data without destination code
merged_passenger_df.destination.isnull().sum()
```

##### There is lots of rows, we need to group them by day and country name

```python

group_data =  merged_passenger_df.groupby(['Country Name', 'day'], as_index = False).count()
#lets see if any null value for destination
group_data.destination.isnull().sum()

#there is zero null value
# as groupby function doesnot include null value
# so we donot need to handle null value manually
```

#### Lets see how a single coutnry data looks like for a month

```python
#lets see the data for a single coutnry/ In This month each day (most probably) occurs once and 
# the number of airplane landed should be shown in the destination column, ignore other columns
group_data[group_data["Country Name"] == "Singapore"].head()
```

```python
#Checksum

#first one printed the occurance of the country name Singapore in the full dataset
# that is this number of plane landed in singapore in this particular month
print(merged_passenger_df[merged_passenger_df["Country Name"] == "Singapore"]["destination"].count())


#second I want to add the number of flight in each day in group data
#both length should be same
print(group_data[group_data["Country Name"] == "Singapore"]["destination"].sum())
```

```python
#Now lets do the same cleaning for all the dataset of 12 months
# same procedure , followed inside the for loop
```

```python
air_data = air_data_list
for i in range(len(air_data)):
    print(air_data[i]+"processing.......")
    data = pd.read_csv(air_data[i])
    merged_passenger_df = data.merge(airport_df, on="destination", how="left")
    group_data = merged_passenger_df.groupby(['Country Name', 'day'], as_index = False).count()
    final_data = group_data[["Country Name", "day", "destination"]]
    final_data.to_csv(air_data[i].replace(".csv.gz", "")+".csv")
    
```

##### Creating a master dataframe for all the months
##### here we basically concatinate all the rows for several month

```python
row = 0
master_df = pd.DataFrame()
for i in range(len(monthly_data)):
    data = pd.read_csv(monthly_data[i])
    df = pd.DataFrame(data)
    master_df = pd.concat([master_df, df], ignore_index=True)
```

```python
master_df.head()
master_df = master_df.iloc[:, 1:]
#master_df.to_csv("cleaned_air_passenger.csv")
```
